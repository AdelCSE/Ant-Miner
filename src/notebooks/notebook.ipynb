```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from src.algorithms.AntMiner.ant_miner import AntMiner
```

```python
df = pd.read_csv('tic-tac-toe.csv')
X = df.drop('class', axis=1)
y = df['class']
```

```python
X
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, shuffle=True, test_size=0.33)
```

```python
antminer = AntMiner()
antminer.fit(X_train, y_train)
```

```python
y_pred = antminer.predict(X_test)
# y_pred to 0 or 1
y_pred = [1 if pred == 'True' else 0 for pred in y_pred]
y_test = [1 if pred == True else 0 for pred in y_test]

print("Accuracy:", accuracy_score(y_test, y_pred))
```

```python
for rule in antminer.discovered_rules:
    rule_str = " AND ".join([f"{term[0]} = {term[1]}" for term in rule[:-1]]) + f" THEN {rule[-1][1]}"
    print(rule_str)
```
