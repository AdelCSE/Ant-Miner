# Ljubljana dataset

```python
from ucimlrepo import fetch_ucirepo 
  
# fetch dataset 
breast_cancer = fetch_ucirepo(id=14) 
  
# data (as pandas dataframes) 
X = breast_cancer.data.features 
y = breast_cancer.data.targets 
  
# metadata 
print(breast_cancer.metadata) 
  
# variable information 
print(breast_cancer.variables)

df = pd.concat([X, y], axis=1)
df
```

```python
df.dtypes
```

```python
# drop duplicates
df = df.drop_duplicates()
```

```python
df.fillna('unknown', inplace=True)
```

```python
missing_values = df.isnull().sum()
missing_values[missing_values > 0]
```

```python
df.describe()
```

```python
df['deg-malig'] = df['deg-malig'].astype(str)
```

```python
# rename Class with class
df.rename(columns={'Class': 'class'}, inplace=True)
```

```python
df['class'] = df['class'].replace({'recurrence-events': 'pos', 'no-recurrence-events': 'neg'})
```

```python
df
```

```python
df.to_csv('/home/adel/Documents/Code/Ant-Miner/datasets/ljubljana.csv', index=False)
```
