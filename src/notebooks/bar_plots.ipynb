```python
# Datasets
datasets = ["TicTacToe-WS", "TicTacToe-TCH",
            "Hepatitis-WS", "Hepatitis-TCH",
            "Ljubljana-WS", "Ljubljana-TCH"]

# Accuracy means
accuracy_without = [72.35, 74.94, 83.16, 83.10, 68.90, 70.13]
accuracy_with    = [81.11, 83.03, 82.06, 82.90, 69.69, 70.02]

# Standard deviations
std_without = [10.60, 8.35, 4.76, 5.20, 3.05, 3.81]
std_with    = [5.40, 7.72, 4.66, 6.39, 7.35, 4.33]

# Convert to numpy arrays
accuracy_without, accuracy_with = np.array(accuracy_without), np.array(accuracy_with)
std_without, std_with = np.array(std_without), np.array(std_with)

# X positions
x = np.arange(len(datasets))
bar_width = 0.25  # match TRR plot

# Colors
colors = ["#000711", "#64a7ce"]

# Plot
fig, ax = plt.subplots(figsize=(10, 5))

# Without pruning
ax.bar(x - bar_width/2, accuracy_without, bar_width,
       yerr=std_without, capsize=3, label="Without Pruning",
       color=colors[0], edgecolor="black", linewidth=1.2, alpha=0.9)

# With pruning
ax.bar(x + bar_width/2, accuracy_with, bar_width,
       yerr=std_with, capsize=3, label="With Pruning",
       color=colors[1], edgecolor="black", linewidth=1.2, alpha=0.9)

# Labels and ticks
ax.set_ylabel("Accuracy (%)", fontsize=13)
ax.set_xlabel("Datasets", fontsize=13)
ax.set_xticks(x)
ax.set_xticklabels(datasets, fontsize=10, fontweight="bold")

# Grid settings
ax.yaxis.grid(True, linestyle=":", linewidth=0.7, alpha=0.8)
ax.xaxis.grid(True, linestyle=":", linewidth=0.7, alpha=0.8)
ax.set_axisbelow(True)

# Legend
ax.legend(frameon=True, edgecolor="black", fontsize=11, fancybox=True)

plt.tight_layout()
plt.show()
```
