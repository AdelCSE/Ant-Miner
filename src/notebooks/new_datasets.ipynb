```python
import pandas as pd
from sklearn.datasets import load_svmlight_file
```

```python
input_file = "C:\\Users\\mokad\\Desktop\\a1a.txt"
output_csv = "C:\\Users\\mokad\\Documents\\Github\\Ant-Miner\\datasets\\SLC\\a1a.csv"
```

```python
X_sparse, y = load_svmlight_file(input_file, n_features=123)

df = pd.DataFrame(X_sparse.todense())
df.insert(0, "class", y)
```

```python
df.columns = ["class"] + [f"attr{i}" for i in range(1, df.shape[1])]
df
```

```python
df['class'] = df['class'].apply(lambda x: 'pos' if x == 1.0 else 'neg')

for col in df.columns:
    if col != 'class':
        df[col] = df[col].apply(lambda x: '1' if x == 1.0 else '0')
```

```python
df
```

```python
for col in df.columns:
    if df[col].nunique() == 1:
        print(f"Column {col} has only one unique value: {df[col].unique()}")
        df.drop(columns=[col], inplace=True)
```

```python
df.to_csv(output_csv, index=False)
```
