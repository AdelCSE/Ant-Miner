```python
import pandas as pd
from scipy.io import arff
```

```python
load_path = "C:\\Users\\mokad\\Desktop\\segment0.dat"
save_path = "C:\\Users\\mokad\\Documents\\Github\\Ant-Miner\\datasets\\SLC"
```

```python
cols = [
    "Region-centroid-col","Region-centroid-row","Region-pixel-count",
    "Short-line-density-5","Short-line-density-2",
    "Vedge-mean","Vegde-sd","Hedge-mean","Hedge-sd",
    "Intensity-mean","Rawred-mean","Rawblue-mean","Rawgreen-mean",
    "Exred-mean","Exblue-mean","Exgreen-mean",
    "Value-mean","Saturatoin-mean","Hue-mean","Class"
]

df = pd.read_csv(
    load_path,
    comment='@',
    header=None,
    names=cols,
    skipinitialspace=True
)
```

```python
df
```

```python
df['Class'] = df['Class'].replace({"negative": 'neg', "positive": 'pos'})
df.rename(columns={"Class": "class"}, inplace=True)
```

```python
# number of unique values per column
for col in df.columns:
    unique_values = df[col].nunique()
    print(f"Column '{col}' has {unique_values} unique values.")
```

```python
df['Short-line-density-5'] = df['Short-line-density-5'].replace({0.000000: 'zeros', 0.111111: 'ones', 0.222222: 'twos', 0.333333: 'threes'})
df['Short-line-density-2'] = df['Short-line-density-2'].replace({0.000000: 'zeros', 0.111111: 'ones', 0.222222: 'twos'})
```

```python
# nb of duplicates
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")

# missing values
missing_values = df.isnull().sum().sum()
print(f"Number of missing values: {missing_values}")
```

```python
# columns with one unique value
one_unique = [col for col in df.columns if df[col].nunique() == 1]
df.drop(columns=one_unique, inplace=True)
print(f"Dropped columns with one unique value: {one_unique}")
```

```python
for col in df.columns:
    if col not in ['class', 'Short-line-density-5', 'Short-line-density-2']:
        df[col] = pd.cut(df[col], bins=20, labels=[str(i) for i in range(20)])
df
```

```python
duplicates = df.duplicated().sum()
print(f"Number of duplicate rows: {duplicates}")
```

```python
df.drop_duplicates(inplace=True)
df.reset_index(drop=True, inplace=True)
df
```

```python
df.to_csv(f"{save_path}/segment0.csv", index=False)
```
